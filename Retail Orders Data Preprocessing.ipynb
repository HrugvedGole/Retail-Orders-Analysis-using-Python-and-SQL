```python
df = pd.read_csv("orders.csv", na_values=["Not Available", "unknown"])
df["Ship Mode"].unique()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

# Converting the order_date from object to datetime 

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%Y-%m-%d")
```

```python
df.dtypes
```

# Renaming columns so they are in the same format

```python
columns = df.columns.str.lower()
columns = columns.str.replace(" ", "_")
columns
```

```python
df.columns = columns
```

```python
df.head()
```

# Deriving New Columns - discount, sale price and profit

```python
df["discount"] = df["list_price"]*df["discount_percent"]*.01
df["discount"]
```

```python
df["sale_price"] = df["list_price"] - df["discount"]
df["sale_price"]
```

```python
df["profit"] = df["sale_price"] - df["cost_price"]
df["profit"]
```

```python
df.head()
```

# Dropping cost price, list price and discount_percent columns

```python
df.drop(columns=["list_price", "cost_price", "discount_percent"], inplace=True)
```

```python
df.head()
```

# Saving the data 

```python
df.to_csv("orders_df.csv", index=False)
```
